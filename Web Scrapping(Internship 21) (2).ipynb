Write a python program to display all the header tags from ‘en.wikipedia.org/wiki/Main_Page’.

```python
page1 = requests.get(' http://en.wikipedia.org/wiki/Main_Page?')
page1
```

```python
soup1 = BeautifulSoup(page1.content)
soup1
```

```python
titles1 = [] 
for i in soup1.find_all('span', class_="mw-headline"):
    titles1.append(i.text) #titles.append(i.text.split('|')[0])
titles1
```
